```python
#Загружаем пакеты из библиотеки
from sympy.plotting import plot
from sympy import Symbol, solve, diff
from sympy import sin, cos, pi, pprint, solveset, S
```

```python
# Определяем переменную
x = Symbol('x')
# Записываем функцию
f = sin(x)*sin(x) - cos(x)*cos(x)
```

```python
# 1. Определить корни
root= solve(f,x)
print(root)
# Переводим ответ в явный числовой формат
for i in root:
  print(i.evalf())
```

```python
# 2. Найти интервалы, на которых функция возрастает
# Находим производную
Proizvod=diff(f,x)
# Находим интервал, где производная > 0 (функция возрастает)
pprint(solveset(Proizvod>0, x, domain = S.Reals), use_unicode = True)
```

```python
# 3. Найти интервалы, на которых функция убывает
# Находим интервал, где производная < 0 (функция возрастает)
pprint(solveset(Proizvod<0, x, domain = S.Reals), use_unicode = True)
```

```python
# 4. Построить график
plot(f)
```

```python
# 5. Вычислить вершину
# Находим производную
Proizvod=diff(f,x)
print(Proizvod)
# Находим х, при котором производная равна 0
solve(Proizvod,x)
```

```python
# 6. Определить промежутки, на котором f > 0
print(solve (f>0, x))
```

```python
# 7. Определить промежутки, на котором f < 0
print(solve(f<0))
```
